## All imports


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

## Utilies loading data classification and kmeans


```python
def _load_data(excel_file_path):
    x = pd.read_excel(excel_file_path,engine='openpyxl')
    print(x.head(5))
    return x

def classify(X_train, X_test, y_train, y_test ,model=LogisticRegression(random_state=42,max_iter=7600)):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy: {} %'.format(accuracy_score(y_test, y_pred) * 100))
    return model

def Kmeans(excel_file_path,n_clusters = 4):
    df = _load_data(excel_file_path)
    kmeans = KMeans(
        init="k-means++",
        n_init=10,
        n_clusters = n_clusters,
        max_iter=300, 
        random_state=42
    )
    scaled_features = scaler.fit_transform(df)
    kmeans.fit(scaled_features)
    return kmeans
    
```

## Evaluting kmeans

```python
kmeans_kwargs = {
        "init": "k-means++",
        "n_init": 10,
        "max_iter": 300,
        "random_state": 42,
   }
df = _load_data("Acviss interview-Training set.xlsx")
```

```python
scaled_features = scaler.fit_transform(df)
```

```python
# Elbow method with scaler
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(
   range(1, 11), sse, curve="convex", direction="decreasing"
)
kl.plot_knee()
```

```python
# without scaler
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(
   range(1, 11), sse, curve="convex", direction="decreasing"
)
kl.plot_knee()
```
