```python
output = circle.copy()
gray = cv2.cvtColor(circle, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
canny = cv2.Canny(gray, 200, 300)
circles_detect = cv2.HoughCircles(canny, cv2.HOUGH_GRADIENT, 1, 20,
              param1=30,
              param2=15,
              minRadius=190,
              maxRadius=200)
```

```python
img = cv2.cvtColor(output, cv2.COLOR_BGR2GRAY)
height,width = img.shape
mask = np.full((height,width), 0, dtype=np.uint8) 
```

```python
xx, yy, radius = None,None,None
if circles_detect is not None:
    # convert the (x, y) coordinates and radius of the circles to integers
    circles_detect = np.uint16(np.around(circles_detect))
    # loop over the (x, y) coordinates and radius of the circles
    for x, y, z in circles_detect[0,:]:
        # draw the circle in the output image, then draw a rectangle
        # corresponding to the center of the circle
        cv2.circle(output,(x,y),z,(0,255,0),10)
        # draw the center of the circle
        cv2.circle(output,(x,y),2,(0,0,255),3)
        cv2.circle(mask,(x,y),z,(255,255,255),thickness=-1)
        xx, yy, radius = x, y ,z
```
